### Q1. What is the difference between a function and a method in Python?
**Answer (detailed):**
- **Function:** A named block of code defined using `def` (or `lambda`) that performs a specific task and can be called directly. Functions are independent of classes/objects.
- **Method:** A function that is defined within a class and is meant to be called on class instances (objects). Instance methods take `self` as the first parameter which refers to the instance.

**Example (both):**


```python
# Function (standalone)
def add(a, b):
    """Return sum of a and b."""
    return a + b

print('Function add(3,5) ->', add(3,5))

# Method (inside a class)
class Calculator:
    def multiply(self, a, b):
        """Instance method: multiply two numbers."""
        return a * b

calc = Calculator()
print('Method calc.multiply(4,2) ->', calc.multiply(4,2))
```

### Q4. What is the purpose of the `return` statement in a Python function?
**Answer:**
- `return` sends a value back to the caller. Without `return`, the function returns `None`.

**Example:**


```python
def add_ret(a,b):
    return a+b

print(add_ret(5,10))

def add_print(a,b):
    print(a+b)

res = add_print(5,10)
print('returned value is', res)
```

### Q5. What are iterators in Python and how do they differ from iterables?
**Answer:**
- An **iterable** is any Python object capable of returning its elements one at a time (has `__iter__`). A **iterator** is the object that does the actual iteration (has `__next__`).
- Iterators maintain state and are consumed.

**Example:**


```python
nums = [10,20,30]
it = iter(nums)
print(next(it))
print(next(it))
print(next(it))
# iterator is now exhausted
```

### Q6. Explain the concept of generators in Python and how they are defined.
**Answer:**
- **Generators** are a simple way to create iterators using `yield`. They pause state between yields and resume on next.

**Example:**


```python
def gen_numbers(n):
    for i in range(n):
        yield i

for val in gen_numbers(3):
    print(val)
```

### Q7. What are the advantages of using generators over regular functions?
**Answer:**
- Advantages: memory efficiency, lazy evaluation, ability to represent infinite streams, and easy pipelining.

**Example (memory):**


```python
def squares_gen(n):
    for i in range(n):
        yield i*i

it = squares_gen(1000000)
for _ in range(5):
    print(next(it))
```

### Q8. What is a lambda function in Python and when is it typically used?
**Answer:**
- **Lambda**: anonymous, single-expression function. Good for short callbacks, `map`, `filter`, `sorted` keys.

**Example:**


```python
pairs = [('b',2), ('a',3), ('c',1)]
print(sorted(pairs, key=lambda p: p[0]))
print(sorted(pairs, key=lambda p: p[1]))
```

### Q9. Explain the purpose and usage of the `map()` function in Python.
**Answer:**
- `map(func, iterable)` applies `func` to every item and returns an iterator of results.

**Example:**


```python
def to_f(c):
    return (c*9/5)+32
print(list(map(to_f, [0,20,30,40])))
```

### Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions?
**Answer:**
- `map` for transformation, `filter` for selection, `reduce` for aggregation.

**Example:**


```python
from functools import reduce
xs = [1,2,3,4,5]
print(list(map(lambda x: x*2, xs)))
print(list(filter(lambda x: x%2==0, xs)))
print(reduce(lambda a,b: a+b, xs))
```

### Q11. Show the internal mechanism of a sum operation using `reduce()` for the list `[47, 11, 42, 13]`.
**Answer:**
- Step-by-step reduce on [47,11,42,13] with combiner f(a,b)=a+b.

**Steps:**
1. 47 + 11 = 58
2. 58 + 42 = 100
3. 100 + 13 = 113

**Example (trace):**


```python
from functools import reduce
nums = [47,11,42,13]
trace = []
def comb(a,b):
    s = a+b
    trace.append((a,b,s))
    return s
res = reduce(comb, nums)
print('trace:', trace)
print('result:', res)
```

### P1. Write a Python function that takes a list of numbers and returns the sum of all even numbers.


```python
def sum_even(nums):
    return sum(x for x in nums if x%2==0)

print(sum_even([1,2,3,4,5,6]))
```

### P2. Create a function that accepts a string and returns the reverse of the string.


```python
def reverse_string(s):
    return s[::-1]

print(reverse_string('Python'))
```

### P3. Write a function that returns the squares of numbers in a list.


```python
def square_list(nums):
    return [x*x for x in nums]

print(square_list([1,2,3,4]))
```

### P4. Write a function to check if a number is prime and print all prime numbers from 1 to 200.


```python
def is_prime(n):
    if n<2:
        return False
    i=2
    while i*i<=n:
        if n%i==0:
            return False
        i+=1
    return True

print([x for x in range(1,201) if is_prime(x)])
```

### P5. Implement an iterator class for the Fibonacci sequence that yields the first `n` numbers.


```python
class Fibonacci:
    def __init__(self,n):
        self.n=n; self.a=0; self.b=1; self.count=0
    def __iter__(self): return self
    def __next__(self):
        if self.count>=self.n:
            raise StopIteration
        self.a, self.b = self.b, self.a+self.b
        self.count += 1
        return self.a

print(list(Fibonacci(10)))
```

### P6. Write a generator function that yields the powers of 2 up to `2^n`.


```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

print(list(powers_of_two(5)))
```

### P7. Write a generator that reads a file line by line and yields each line without the newline character.


```python
# Generator function to read file line by line
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()   # yield removes '\n' at the end of each line

# Example usage
# First, let's create a sample text file
with open("sample.txt", "w") as f:
    f.write("Hello World\n")
    f.write("This is line 2\n")
    f.write("This is line 3\n")

# Using the generator
for line in read_file_line_by_line("sample.txt"):
    print(line)

```

### P10. Remove vowels from a string using `filter()`.


```python
def remove_vowels(s):
    return ''.join(filter(lambda ch: ch.lower() not in 'aeiou', s))

print(remove_vowels('Hello World'))
```
